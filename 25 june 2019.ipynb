```python
#Function to generate the marks file for n students

from random import randint

def generateMarks(n,lb,ub):
    with open('files/marks.txt','w') as f:
        for i in range(0,n):
            r = randint(lb,ub)
            f.write(str(r)+'\n')
    return
generateMarks(3000,0,100)
```
