```python
#linear graph 

import matplotlib.pyplot as pit
pit.title("Basic linear Graph")
pit.plot([1,2,3],[1,2,3])
pit.show()
```

```python
# Graph to plot distance and time

distance = list(range(2,11,2))
time = list(range(1,6))

#d=2*t
pit.title("Distance Vs Time")
pit.xlabel("Time")
pit.ylabel("Distance")
pit.grid()
pit.plot(time,distance,"bo")
pit.show()
```

```python
# Mapping between time and velocity
# v=a t
t = list(range(1,11))
a = list(range(1,len(t)+1))
#v = [2 * i for i in t]
v=[a[i] * i for i in range(len(t))]
pit.plot(t,v,"yo")
pit.grid()
pit.show()
```

```python
n = list(range(10))
sqn= [i**3 for i in  n]
pit.plot(n,sqn)
pit.grid()
pit.show()
```

```python
# Voltage - Time
import numpy as np
#t = [i/10 for i in range(1,21)]
t = np.arange(0.1,2.0,0.1)
v = [1 + np.sin(2*np.pi*i) for i in t]
pit.title("Time VS Voltage Graph")
pit.plot(t,v)
pit.grid()
pit.xlabel("Time")
pit.ylabel("Voltage")


pit.savefig("images/Voltage-Time.png")
pit.show()
```

```python
# plot Income of Arkansas from 2005 to 2013
import pandas as pd
df = pd.read_csv('files/income.csv')
years = df.columns[2:]
incomeArkanasas = df.values[3, 2:]

#incomeArkanasas

pit.title("Income of Arkanasas")
pit.xlabel("Year")
pit.ylabel("Income")
pit.grid()
pit.plot(years,incomeArkanasas)
pit.show()



```

```python
df

incomeallstates = []
for i in df.columns[2:]:
    incomeallstates.append(df[i].mean())
pit.plot(years,incomeallstates)
pit.title("All States Income")
pit.xlabel("Year")
pit.ylabel("Income")
pit.grid()
pit.show()
    
```
